```python
from rgrow import State
from rgrow.kblock import KBlock, KBlockParams, KBlockTile
```

```python
params = KBlockParams(
    [
        KBlockTile("S", 40e-9, ["null", "rb", "bb", "null"], "purple"),
        KBlockTile("RB", 40e-9, ["null", "rb", "e1", "rb*"], "red"),
        KBlockTile("BB", 100e-9, ["bb*", "e1", "bb", "null"], "blue"),
        KBlockTile("00", 100e-9, ["e0*", "e0", "e0", "e0*"], "white"),
        KBlockTile("10", 100e-9, ["e1*", "e1", "e1", "e0*"], "green"),
        KBlockTile("01", 100e-9, ["e0*", "e1", "e1", "e1*"], "yellow"),
        KBlockTile("11", 100e-9, ["e1*", "e0", "e0", "e1*"], "orange"),
    ],
    {"e1": 1000e-9, "e0": 1000e-9},
    seed={
        (4,4): "S"
    },
    binding_strength={
        "rb": "ACTGCTAGTACGTAGCGTAA",
        "bb": "ACTGCTAGTACGTAGCGTAA",
        "e0": "ACTGCTAGTA",
        "e1": "ACTGCTAGTA",
    },
    temp=47
)
```

```python
sys = KBlock(params)
state = State((32, 32), "square", "none")
sys.setup_state(state)
sys.update_state(state)
sys.evolve(state, show_window=True)
```
